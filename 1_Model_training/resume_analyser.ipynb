```python
import re
cleaned_resumes = []

for resume in df['Resume_str']:
    lis = resume.replace(",", "").replace("/", "").replace('.', '').replace('(', '').replace('-', '').replace(')', '').replace('_', '').lower().split()
    lis = [word for word in lis if not word.isdigit()]
    obj = list(dict.fromkeys(lis))
    jo = " ".join(obj)
    cleaned_resumes.append(jo)

df['Resume_str'] = cleaned_resumes
```

```python
import nltk   # remove the like loved, loving to love only so not repeat the same word using prefix or suffix

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y) 
```

```python
import re
data = []
for text in df['Resume_str']:
    clean_text = re.sub(r'[^a-zA-Z\s]', '', text)
    data.append(clean_text)

df['Resume_str'] = data    
```

```python
lis = []
for cat in df['Category']:
    str1 = cat.replace("-", "").replace(" ", "").lower()
    lis.append(str1)

df['Category'] = lis
    
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['Resume_str'], df['Category'], test_size=0.2, random_state=42, stratify=df['Category'])

X_train.shape, X_test.shape
```

```python
X_train
```

```python
vectorizer = TfidfVectorizer(stop_words = 'english',ngram_range=(1,2),
    min_df=2,
    max_df=0.95)

```

```python
X_train_new = vectorizer.fit_transform(X_train)
X_test_new = vectorizer.transform(X_test)
```

```python
X_train_new.shape
```

```python
lb = LabelEncoder()
y_train_new = lb.fit_transform(y_train)
y_test_new = lb.transform(y_test)
```

```python
y_train_new.shape
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_new, y_train_new)

y_pred = lr.predict(X_test_new)

accuracy_score(y_test_new, y_pred)
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train_new, y_train_new)

y_pred = dtc.predict(X_test_new)

accuracy_score(y_test_new, y_pred)
```

```python
from sklearn.svm import LinearSVC

model = LinearSVC()

model.fit(X_train_new, y_train_new)

y_pred = model.predict(X_test_new)
accuracy_score(y_test_new, y_pred)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()

model.fit(X_train_new, y_train_new)

y_pred = model.predict(X_test_new)
accuracy_score(y_test_new, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)

model.fit(X_train_new, y_train_new)

y_pred = model.predict(X_test_new)
accuracy_score(y_test_new, y_pred)
```

```python
from sklearn.linear_model import SGDClassifier

model = SGDClassifier(random_state=42)

model.fit(X_train_new, y_train_new)

y_pred = model.predict(X_test_new)
accuracy_score(y_test_new, y_pred)
```

```python
from xgboost import XGBClassifier

model = XGBClassifier()

model.fit(X_train_new, y_train_new)

y_pred = model.predict(X_test_new)
accuracy_score(y_test_new, y_pred)
```
