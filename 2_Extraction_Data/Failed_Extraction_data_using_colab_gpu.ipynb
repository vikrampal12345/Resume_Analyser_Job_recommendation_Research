```python
import os
import re
import cv2
import easyocr
import pandas as pd

from tqdm import tqdm
```

```python
import torch

print(torch.cuda.is_available())
print(torch.cuda.get_device_name(0))
```

```python
reader = easyocr.Reader(
    ['en'],
    gpu=True
)
```

```python
dataset_path="/root/.cache/kagglehub/datasets/youssefkhalil/resumes-images-datasets"

image_paths=[]

for root,dirs,files in os.walk(dataset_path):

    for file in files:

        if file.lower().endswith((".jpg",".jpeg",".png")):

            image_paths.append(os.path.join(root,file))

print("Total Images :",len(image_paths))
```

```python
image=image_paths[0]

result=reader.readtext(
    image,
    detail=0,
    paragraph=True
)

text="\n".join(result)

print(text)
```

```python
def clean_text(text):

    text = re.sub(r"\s+", " ", text)

    text = re.sub(r"[^\x00-\x7F]+", " ", text)

    text = text.strip()

    return text
```

```python
data = []

for image_path in tqdm(image_paths):

    try:

        result = reader.readtext(
            image_path,
            detail=0,
            paragraph=True
        )

        text = " ".join(result)

        text = clean_text(text)

        data.append({

            "filename": os.path.basename(image_path),

            "image_path": image_path,

            "resume_text": text

        })

    except Exception as e:

        print(f"Error : {image_path}")
```

```python
len(data)
```

```python
df = pd.DataFrame(data)

print(df.shape)

df.head()
```

```python
df.to_csv(
    "resume_text_dataset.csv",
    index=False,
    encoding="utf-8"
)

print("CSV Saved Successfully!")
```
